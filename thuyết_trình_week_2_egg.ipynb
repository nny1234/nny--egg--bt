**1. EDA là gì**

EDA (Exploratory Data Analysis – Phân tích Khám phá Dữ liệu) là một bước quan trọng trước khi làm bất kỳ một bài toán ML với dữ liệu dạng bảng nào.

Trước khi xây dựng mô hình, bạn cần xây dựng đặc trưng. Trước khi xây dựng đặc trưng, bạn phải làm bước khám phá dữ liệu.

Mục đích của EDA

Bước EDA này giúp chúng ta có cái nhìn đầu tiên về dữ liệu. Bạn cần có một cảm giác nhất định về những gì mình có trong tay trước khi có những chiến lược xây dựng mô hình. EDA giúp bạn mường tượng được độ phức tạp của bài toán và vạch ra những bước đầu tiên cần làm.

Việc khám phá dữ liệu không chỉ dừng lại ở lần đầu tiên trước khi xây dựng đặc trưng mà còn cần được thực hiện trong suốt quá trình phát triển hệ thống. Sau khi xây dựng xong các đặc trưng, bạn cũng cần làm lại EDA một lần nữa để xem dữ liệu đã qua xử lý đó đã thực sự sạch chưa. Ngoài ra, sau khi xây dựng và phân tích mô hình, ta cũng thường xuyên cần quay lại EDA để tiếp tục khám phá những điều còn ẩn giấu trong dữ liệu bài toán. Càng hiểu sâu về dữ liệu, bạn sẽ càng sớm giải thích được những hành vi của mô hình và đưa ra những thay đổi phù hợp.

**Note:** Để phân biệt dữ liệu trước và sau bước tiền xử lý, các cột trong bảng dữ liệu ban đầu được gọi là trường thông tin, trường dữ liệu hoặc đôi khi ngắn gọn là trường. Các cột đã được xử lý và sẵn sàng cho việc huấn luyện mô hình được gọi là đặc trưng.

**Kiểu dữ liệu của mỗi trường**


Việc kiểm tra kiểu dữ liệu của mỗi cột (column) trong tập dữ liệu (DataFrame) có tác dụng quan trọng trong việc:

Kiểm tra đúng định dạng dữ liệu: Bằng cách kiểm tra kiểu dữ liệu của mỗi cột, bạn có thể đảm bảo rằng dữ liệu được đúng định dạng mong muốn. Ví dụ, nếu bạn mong đợi một cột chứa dữ liệu số nguyên, việc kiểm tra kiểu dữ liệu sẽ giúp bạn xác minh xem cột đó thực sự chứa kiểu dữ liệu số nguyên hay không.

Xác định loại xử lý phù hợp: Dựa trên kiểu dữ liệu của mỗi cột, bạn có thể xác định loại xử lý phù hợp cho từng cột. Ví dụ, nếu một cột có kiểu dữ liệu ngày tháng, bạn có thể áp dụng các phép toán và chức năng liên quan đến ngày tháng để phân tích dữ liệu hoặc tính toán.

Phát hiện lỗi hoặc không chính xác: Kiểm tra kiểu dữ liệu cũng có thể giúp bạn phát hiện các lỗi hoặc không chính xác trong dữ liệu. Nếu một cột được giả định là kiểu dữ liệu số, nhưng thực tế lại chứa các giá trị không phải số, việc kiểm tra kiểu dữ liệu sẽ giúp bạn phát hiện lỗi này.

Truy cập và xử lý dữ liệu dễ dàng: Biết kiểu dữ liệu của mỗi cột cũng giúp bạn truy cập và xử lý dữ liệu dễ dàng hơn. Bạn có thể áp dụng các phương thức và chức năng phù hợp với kiểu dữ liệu để thực hiện các thao tác cần thiết trên cột đó.

Để kiểm tra kiểu dữ liệu của mỗi cột trong DataFrame, bạn có thể sử dụng phương thức .dtypes hoặc .info() của DataFrame.

"Non-Null Count" là số lượng giá trị không null (không rỗng) trong mỗi cột của một DataFrame. Khi thực hiện phương thức .info() trên DataFrame, thông tin được hiển thị bao gồm "Non-Null Count" cho mỗi cột.

Số lượng "Non-Null Count" cho mỗi cột cho biết có bao nhiêu giá trị không bị thiếu (null) trong cột đó. Các giá trị null thường xuất hiện khi dữ liệu bị thiếu hoặc chưa được cung cấp. Việc kiểm tra "Non-Null Count" giúp xác định mức độ hoàn chỉnh của dữ liệu trong mỗi cột.

Thông thường, một cột với "Non-Null Count" lớn gần bằng số lượng hàng (rows) trong DataFrame cho thấy dữ liệu trong cột đó khá đầy đủ. Ngược lại, nếu một cột có số lượng "Non-Null Count" thấp hơn nhiều so với số lượng hàng, có thể cho thấy có sự thiếu sót hoặc lỗ hổng trong dữ liệu của cột đó.

Việc kiểm tra số lượng "Non-Null Count" trong các cột của DataFrame giúp định rõ tình trạng dữ liệu và có thể cung cấp thông tin về khả năng xử lý, phân tích và điền giá trị cho các giá trị thiếu trong dữ liệu.


Numerical_stats là một thuật ngữ trong phân tích dữ liệu để chỉ các thống kê cơ bản được tính toán trên các numerical_features trong tập dữ liệu. Các numerical_stats cung cấp thông tin quan trọng về tính tập trung, phân tán và phân bố của dữ liệu số.

Khi ta có các numerical_stats, ta biết được các thông số cơ bản như sau:

Giá trị trung bình (mean): Đây là giá trị trung bình của các numerical_features, thể hiện giá trị trung bình của dữ liệu.

Độ lệch chuẩn (standard deviation): Đây là một đo lường về mức độ phân tán của dữ liệu số, cho biết mức độ biến động của dữ liệu quanh giá trị trung bình.

Giá trị tối thiểu (minimum) và giá trị tối đa (maximum): Đây là giá trị nhỏ nhất và lớn nhất trong các numerical_features, cho biết phạm vi giá trị của dữ liệu.

Phần centile (percentile): Đây là các giá trị phân vị của dữ liệu, cho biết giá trị nằm trong phần trăm nhất định của phân bố dữ liệu.

Độ méo (skewness) và độ nhọn (kurtosis): Đây là các độ đo về hình dạng phân phối của dữ liệu. Độ méo đo độ lệch của phân phối so với phân phối chuẩn, trong khi độ nhọn đo độ tập trung của dữ liệu xung quanh giá trị trung bình.

Các numerical_stats cho ta cái nhìn tổng quan về tính chất của dữ liệu số trong tập dữ liệu, giúp ta hiểu về mức độ biến động, phân bố và hình dạng của dữ liệu số. Điều này rất hữu ích trong việc đưa ra quyết định phân tích và phân tích sâu hơn về dữ liệu.

```python
#12
from scipy.stats import skew, kurtosis
sale_price_skewness = skew(df_selected["SalePrice"])
sale_price_kurtosis = kurtosis(df_selected["SalePrice"])

print("Skewness of SalePrice:", sale_price_skewness)
print("Kurtosis of SalePrice:", sale_price_kurtosis)
```

Skewness và Kurtosis là hai khái niệm thống kê được sử dụng để mô tả hình dạng của phân phối dữ liệu.

1. Skewness (độ lệch): Độ lệch đo lường độ chênh lệch của phân phối dữ liệu so với phân phối chuẩn (phân phối đối xứng). Giá trị skewness có thể là dương, âm hoặc gần bằng 0.

- Skewness dương (+): Cho thấy dữ liệu có xu hướng lệch về phía phải, tức là có đuôi dài hơn về bên phải của phân phối.

- Skewness âm (-): Cho thấy dữ liệu có xu hướng lệch về phía trái, tức là có đuôi dài hơn về bên trái của phân phối.

- Skewness gần bằng 0: Cho thấy dữ liệu có phân phối gần đối xứng.

2. Kurtosis (độ nhọn): Kurtosis đo lường độ tập trung của dữ liệu tại đỉnh phân phối và hình dạng đuôi của phân phối so với phân phối chuẩn. Giá trị kurtosis có thể là lớn hơn, nhỏ hơn hoặc bằng 0.

- Kurtosis lớn hơn 0: Cho thấy dữ liệu có đỉnh cao hơn và đuôi dày hơn so với phân phối chuẩn, có xu hướng tập trung hơn tại đỉnh.

- Kurtosis nhỏ hơn 0: Cho thấy dữ liệu có đỉnh thấp hơn và đuôi mỏng hơn so với phân phối chuẩn, có xu hướng phân tán hơn.

- Kurtosis bằng 0: Cho thấy dữ liệu có phân phối tương đối chuẩn.

```python
#16
# Tạo cột Season dựa trên cột MoSold
df['Season'] = pd.cut(df['MoSold'], bins=[0, 3, 6, 9, 12], labels=['Spring', 'Summer', 'Fall', 'Winter'])
```

```python
# Tạo hàm ánh xạ tháng sang mùa
def map_month_to_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

# Áp dụng hàm ánh xạ vào cột MoSold để tạo cột Season
df['Season'] = df['MoSold'].apply(map_month_to_season)
```

```python
from scipy.stats import f_oneway

# Kiểm định one-way ANOVA
p_value = f_oneway(df[df['Season'] == 'Spring']['SalePrice'],
                   df[df['Season'] == 'Summer']['SalePrice'],
                   df[df['Season'] == 'Fall']['SalePrice'],
                   df[df['Season'] == 'Winter']['SalePrice']).pvalue

if p_value < 0.05:
    print("Có sự khác biệt đáng kể ")
else:
    print("K có sự khác biệt đáng kể ")
```

Phương pháp kiểm định thống kê one-way ANOVA (Analysis of Variance) được sử dụng để kiểm tra sự khác biệt về giá trị trung bình của một biến phụ thuộc (như giá bán) giữa ba hoặc nhiều nhóm độc lập (trong trường hợp này là các mùa: Xuân, Hè, Thu, Đông). Ý tưởng của phương pháp này là so sánh phương sai giữa các nhóm và phương sai trong nhóm để xác định xem sự khác biệt giữa các nhóm có ý nghĩa thống kê hay không.
